# 1. Which movies are with highest and lowest Budget?

```python
data.columns
```

```python
top_budget = data[(data.budget == data['budget'].max())|(data.budget == data['budget'].min())]
sns.barplot(top_budget['original_title'], top_budget['budget'], palette = 'Paired')
plt.title('Highest & Lowest Budgetary Movies', fontsize = 14)
plt.xlabel('Movie Title')
plt.ylabel('Budget')
warnings.simplefilter(action="ignore", category=FutureWarning)
plt.show()
```

```python
movie_and_budget = data[['original_title','budget']]
sns.barplot( movie_and_budget.sort_values(by='budget', ascending=False).head(10).budget,
              movie_and_budget.sort_values(by='budget', ascending=False).head(10).original_title, palette = 'viridis')

plt.title('Top 10 movies based on their Budget', fontsize = 16)
plt.xlabel('Budget', fontsize = 14)
plt.ylabel('Movie Title', fontsize = 14)

plt.show()
```

"The Warrior's way" is the highest budgetary movie, let's see if it made a high Revenue & Profit then..

# 3. Is there a relationship between The Revenue & The Budget?

```python
sns.scatterplot(x = data['revenue'], y = data['budget'], color='#A52A2A')

plt.title('Revenue & Budget', fontsize=14)
plt.xlabel('Revenue')
plt.ylabel('Budget')

data_corr = data.corr()
print("Correlation Between Revenue And Budget : ",data_corr.loc['revenue','budget'])
```

There's a positive relationship between Revenues & Budget by 0.73, which means there's
a high chance that movies with higher budget make high revenues.

# 7. Classification of Movies Based on Votes

```python
def votes(votes): 
    if votes >= 7.0:
        return 'Excellent'
    elif votes >= 6.0:
        return 'Average'
    else:
        return 'Poor'
```

```python
data['vote_class'] = data['vote_average'].apply(votes)
```

# 8. Which Genre Has The Highest Release Of Movies?

```python
genre = data['genres'].str.get_dummies(sep= '|')
genre_1 = genre[genre.columns].apply(lambda x: sum(x.values))
genre_1.sort_values(0, ascending = False).head(14).plot.pie(figsize = (10,10), fontsize = 15, autopct= "%1.1f%%")
plt.title('Genres Releases in %', fontsize = 20)
plt.show()
```

Observed that Drama genre has the highest release of movies followed by Comedy and Thriller.
